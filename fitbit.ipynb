## Loading Data

```python
daily_activity = pd.read_csv("mturkfitbit_export_3.12.16-4.11.16/Fitabase Data 3.12.16-4.11.16/dailyActivity_merged.csv")
heart_rate = pd.read_csv("mturkfitbit_export_3.12.16-4.11.16/Fitabase Data 3.12.16-4.11.16/heartrate_seconds_merged.csv")
calories = pd.read_csv("mturkfitbit_export_3.12.16-4.11.16/Fitabase Data 3.12.16-4.11.16/hourlyCalories_merged.csv")
intensities = pd.read_csv("mturkfitbit_export_3.12.16-4.11.16/Fitabase Data 3.12.16-4.11.16/hourlyIntensities_merged.csv")
steps = pd.read_csv("mturkfitbit_export_3.12.16-4.11.16/Fitabase Data 3.12.16-4.11.16/hourlySteps_merged.csv")
calories_narrow = pd.read_csv("mturkfitbit_export_3.12.16-4.11.16/Fitabase Data 3.12.16-4.11.16/minuteCaloriesNarrow_merged.csv")
intensities_narrow = pd.read_csv("mturkfitbit_export_3.12.16-4.11.16/Fitabase Data 3.12.16-4.11.16/minuteIntensitiesNarrow_merged.csv")
mets_narrow = pd.read_csv("mturkfitbit_export_3.12.16-4.11.16/Fitabase Data 3.12.16-4.11.16/minuteMETsNarrow_merged.csv")
sleep = pd.read_csv("mturkfitbit_export_3.12.16-4.11.16/Fitabase Data 3.12.16-4.11.16/minuteSleep_merged.csv")
steps_narrow = pd.read_csv("mturkfitbit_export_3.12.16-4.11.16/Fitabase Data 3.12.16-4.11.16/minuteStepsNarrow_merged.csv")
weight = pd.read_csv("mturkfitbit_export_3.12.16-4.11.16/Fitabase Data 3.12.16-4.11.16/weightLogInfo_merged.csv")
```

```python
# Checking for null values
dataset_list = [(daily_activity, "daily_activity"), (heart_rate, "heart_rate"), (calories, "calories"), (intensities, "intensities"), (steps, "steps"), 
                (calories_narrow, "calories_narrow"), (intensities_narrow, "intensities_narrow"), (mets_narrow, "mets_narrow"), (sleep, "sleep"), 
                (steps_narrow, "steps_narrow"), (weight, "weight")]
for dataset, name in dataset_list:
    null_vals = dataset.isnull().values.any()
    print(f"{name}: {null_vals}")
```

```python
# Remove column that has null since almost all data is null in 'Fat' column
weight = weight.drop('Fat', axis=1)
display(weight.head())
```

```python
# Merging datasets: weightpounds, bmi, daily activity, heart_rate (find a way to aggregate), calories (aggregate), steps, sleep_value (target)
merged_df = pd.merge(daily_activity, weight[['Id', 'WeightPounds', 'BMI']], on = 'Id', how='left')
display(merged_df.head())
```

```python
# Merging heart rate
# Convert 'time' to date
heart_rate['Time'] = pd.to_datetime(heart_rate['Time'], format = '%m/%d/%Y %I:%M:%S %p')

# Extract 'date' part from 'time'
heart_rate['Date'] = heart_rate['Time'].dt.date

# Convert 'ActivityDate' to datetime
merged_df['ActivityDate'] = pd.to_datetime(merged_df['ActivityDate'], format = '%m/%d/%Y')

# Extract date part 
merged_df['Date'] = merged_df['ActivityDate'].dt.date
```

```python
# Aggregate Heart Rate data to daily activity 

heart_rate_summary = heart_rate.groupby(['Id', 'Date'])['Value'].agg(
    mean_heart_rate='mean',
    max_heart_rate='max',
    min_heart_rate='min',
    std_heart_rate='std'
).reset_index()
```

```python
# Merging heart rate to merged_df
merged_df = pd.merge(merged_df, heart_rate_summary[['Id', 'Date','mean_heart_rate']], on=['Id', 'Date'], how='left')
```

```python
# Merging sleep 

# Convert 'date' to datetime
sleep['date'] = pd.to_datetime(sleep['date'], format='%m/%d/%Y %I:%M:%S %p')

# Extract date part from 'date'
sleep['Date'] = sleep['date'].dt.date

# Create indicator columns for sleep stages
sleep['asleep'] = (sleep['value'] == 1).astype(int)
sleep['restless'] = (sleep['value'] == 2).astype(int)
sleep['awake'] = (sleep['value'] == 3).astype(int)

# Aggregate sleep data
sleep_summary = sleep.groupby(['Id', 'Date']).agg(
    total_minutes_asleep=('asleep', 'sum'),
    total_minutes_restless=('restless', 'sum'),
    total_minutes_awake=('awake', 'sum'),
    sleep_episodes=('logId', pd.Series.nunique)
).reset_index()

# Total sleep minutes
sleep_summary['total_sleep_minutes'] = sleep_summary['total_minutes_asleep']

# Total time in bed
sleep_summary['total_time_in_bed'] = sleep_summary['total_sleep_minutes']  + sleep_summary['total_minutes_restless']+ sleep_summary['total_minutes_awake']

# Sleep Efficiency
sleep_summary['sleep_efficiency'] = (sleep_summary['total_sleep_minutes'] / sleep_summary['total_time_in_bed']) * 100

# Merge the sleep summary with merged_df
merged_df = pd.merge(merged_df, sleep_summary, on=['Id', 'Date'], how='left')


```

```python
# Assuming your DataFrame is named 'merged_df'
nan_count = merged_df['mean_heart_rate'].isnull().sum()
print(f"Number of NaN values in 'mean_heart_rate' column: {nan_count}")

nan_count = merged_df['WeightPounds'].isnull().sum()
print(f"Number of NaN values in 'WeightPounds' column: {nan_count}")

nan_count = merged_df['BMI'].isnull().sum()
print(f"Number of NaN values in 'BMI' column: {nan_count}")

nan_count = merged_df['sleep_efficiency'].isnull().sum()
print(f"Number of NaN values in 'Sleep Efficiency' column: {nan_count}")
```

```python
# Filter rows where all specified columns are NaN
nan_rows = merged_df[
    merged_df['WeightPounds'].isnull() |
    merged_df['BMI'].isnull() |
    merged_df['mean_heart_rate'].isnull() | 
    merged_df['sleep_efficiency'].isnull()
]

# Print the filtered rows
display(nan_rows)
```
